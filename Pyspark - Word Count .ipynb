# Word Count using Spark Resilient Distributed Datasets (RDD) 

Load text file in spark

```python
rdd = spark.sparkContext.textFile("file:///home/hadoop/Downloads/shakespeare.txt")

# for element in rdd.collect():
#     print(element)
```

Perform word count alogorithm on the Shakespeare text loaded above

```python
# Flatmap 
rdd2=rdd.flatMap(lambda x: x.split(" "))

# map
rdd3=rdd2.map(lambda x: (x,1))
    
# reduceByKey
rdd4 = rdd3.reduceByKey(lambda a,b: a+b)
for element in rdd4.collect():
    print(element)
```
